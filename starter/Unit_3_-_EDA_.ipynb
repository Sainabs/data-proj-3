### Step 5 - Create Plots for Each Variable in Dataset

### Plot for 'age' column

```python
# create plot for age column
```

```python
plt.figure(figsize=(8,6))

ax = sns.histplot(data=df, x='age', palette='magma', bins=8)

plt.xlabel("Age", fontsize=14)
plt.ylabel("Number of Managers", fontsize=14)

plt.title("Distribution of Manager Ages in Survey", fontsize=15);
```

### Plot for 'industry' column

```python
# create plot for industry column
plt.figure(figsize=(8,8))

data = df['industry'].value_counts()

labels = ["Healthcare", "Technology", "Finance", "Business"]

# colors = sns.color_palette('Set2')

plt.title("% Of Managers In Each Industry", fontsize=16)

ax = plt.pie(data, labels = labels, autopct = '%.0f%%');
```

The manager are equally divided in each industry they manage.

The density of manager salaries is mostly on 100,000 dollars area. So the salary of most manager is 100,000 dollars.

- mean
> The mean is the average of a data set.
- median
> the value or quantity lying at the midpoint of a frequency distribution of observed values or quantities.
- mode
> The mode is the value that appears most often in a set of data. 
- central tendency
> A measure of central tendency is a single value that attempts to describe a set of data by identifying the central position within that set of data.
- distribution
> A statistical distribution, or probability distribution, describes how values are distributed for a field. In other words, the statistical distribution shows which values are common and uncommon.
- variance
> The term variance refers to a statistical measurement of the spread between numbers in a data set.
- standard deviation
> Standard deviation is a statistic that measures the dispersion of a dataset relative to its mean and is calculated as the square root of the variance. The standard deviation is calculated as the square root of variance by determining each data point's deviation relative to the mean.

#### Fill in blanks with brief description of the following terms:

- **Hypothesis**
> explanation made on the basis of limited evidence as a starting point for further investigation.

- **Null Hypothesis**
> no significant statistical difference between specified populations, any observed difference being due to sampling or experimental error.

- **P-Value**
> the probability that a particular statistical measure, such as the mean or standard deviation, of an assumed probability distribution will be greater than or equal to (or less than or equal to) observed results.
P > 0.05 is the probability that the null hypothesis is true.

### What is the correlation between salary and education?

#### Your Hypothesis 
>The higher the education the higher the salary

```python
# calculate correlation and pvalue of salary and education with spearmanr method
spearmanr(df['salary'], df['education'])
```

There is positive correlation between salary and education. 

#### Are these results statistically significant? How do you know?

yes, there is significant statistical difference as the p value is 0.

### What is the correlation between salary and age?

#### Your Hypothesis: 
>The older the age the higher the salary.

```python
# calculate correlation and pvalue of salary and age with pearsonr method
pearsonr(df['salary'], df['age'])
```

There is positive correlation between salary and age. 

#### Are these results statistically significant? How do you know? 

yes, there is significant statistical difference as the p value is 0.

### What is the correlation between salary and industry experience?

#### Crosstab of average salary grouped by education and industry

```python
# use pd.crosstab() to calculate mean salary in each industry grouped by education level
pd.crosstab(index=df['industry'], columns=df['education'], values=df['salary'], aggfunc=np.mean).round(0)
```

#### Heatmap of crosstab of average salary grouped by education and industry

```python
# save above result to new variable
hm_ct = pd.crosstab(index=df['industry'], columns=df['education'], values=df['salary'], aggfunc=np.mean).round(0)
```

```python
# create heatmap with annotated mean salary in each industry and education level

ax = sns.heatmap(hm_ct, annot=True, cmap="PiYG", linewidths=.5, fmt='g', xticklabels=['High School', 'Some College', "Bachelor's Degree", "Master's Degree"])

plt.xlabel("Education Level", fontsize=14)
plt.ylabel("Industry", fontsize=14)

plt.title("Manager Salaries Across Industries by Education Level", fontsize=18);
```

##### Insights:

Managers with bachelors education working in the technology industry have the highest salaries.
Managers with some college education working in the healthcare industry have the lowest salaries.
